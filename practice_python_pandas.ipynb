# Merging

```python
import pandas as pd

df1=pd.DataFrame({'HPI':[80,90,70,60], "Interest Rate":[2,1,2,3], "IND_GDP":[50,45,45,67]}, index=[2001,2002,2003,2004])

df2=pd.DataFrame({'HPI':[80,90,70,60], "Interest Rate":[2,1,2,3], "IND_GDP":[50,45,45,67]}, index=[2005,2006,2007,2008])

merge=pd.merge(df1,df2,on="HPI")

print(merge)
```

# Joining

```python
import pandas as pd

df1=pd.DataFrame({"Interest Rate":[2,1,2,3], "IND_GDP":[50,45,45,67]}, index=[2001,2002,2003,2004])

df2=pd.DataFrame({"Low_tier_HPI":[50,45,37,64], "Unemployment":[1,3,5,6]}, index=[2001,2003,2004,2005])

joined=df1.join(df2)

print(joined)
```

# Concatenation

```python
import pandas as pd

df1=pd.DataFrame({'HPI':[80,90,70,60], "Interest Rate":[2,1,2,3], "IND_GDP":[50,45,45,67]}, index=[2001,2002,2003,2004])

df2=pd.DataFrame({'HPI':[80,90,70,60], "Interest Rate":[2,1,2,3], "IND_GDP":[50,45,45,67]}, index=[2005,2006,2007,2008])

concat=pd.concat([df1,df2])

print(concat)
```

# Integral FUnctions

```python
from scipy import integrate
help(integrate.quad)
```

```python
import scipy
i=scipy.integrate.quad(lambda x:special.exp10(x),0,1)
print(i)
```

```python
e=lambda x,y:x*y**2
f=lambda x:1
g=lambda x:-1
integrate.dblquad(e,0,2,f,g)
```

# Fourier Transformations

```python
from scipy.fftpack import fft,ifft
import numpy as np
x=np.array([1,2,3,4])
y=fft(x)
z=ifft(x) # Inverse fourier transform
print(y)
print(z)
```

# Linear Algebra

```python
from scipy import linalg
import numpy as np
a=np.array([[1,2],[3,4]])
b=linalg.inv(a)
print(b)
```

# Interpolation Function

```python
import matplotlib.pyplot as plt
from scipy import interpolate
x=np.arange(5,20)
y=np.exp(x/3.0)
f=interpolate.interp1d(x,y)
x1=np.arange(6,12)
y1=f(x1) # use interploation function returned by interp1d
plt.plot(x,y,'o', x1, y1, '--')
plt.show()
```
